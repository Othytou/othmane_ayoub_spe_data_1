## `0_scraping`

### Read the Dataframe with  coordinates

```python
import pandas as pd
import requests
from io import BytesIO

# URL of the pickled DataFrame in the GitHub repository
url = "https://raw.githubusercontent.com/Selimmmm/spe1/fff081c86c6c17ae30b96f0f250f2c90cabfd09c/projets/projet_III/data/df_with_coordinates_str.pk"

# Make a request to get the pickled file
response = requests.get(url)

# Convert the binary response content to a BytesIO object
data = BytesIO(response.content)

# Use pandas to read the pickled DataFrame from the BytesIO object
df = pd.read_pickle(data)
df
```
